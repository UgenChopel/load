## Overall load profile (with industrial load)

```python
df_baseLoad = pd.concat([df1, df3, df4, df5, df6, df8, df10, df11, df12, df14, dfI1, dfI3, dfI7, dfI9, dfI10, dfI12, dfI13, dfI14], axis=1, join='inner').sort_index()

df_baseLoad['Total_P'] = df1['P1'] + df3['P3'] + df4['P4'] + df5['P5'] + df6['P6'] + df8['P8'] + df10['P10'] + df11['P11'] + df12['P12'] + df14['P14'] + dfI1['PI_1'] + dfI3['PI_3'] + dfI7['PI_7']+ dfI9['PI_9'] + dfI10['PI_10']+ dfI12['PI_12']+ dfI13['PI_13']+ dfI14['PI_14']
df_baseLoad['Total_Q'] = df1['Q1'] + df3['Q3'] + df4['Q4'] + df5['Q5'] + df6['Q6'] + df8['Q8'] + df10['Q10'] + df11['Q11'] + df12['Q12'] + df14['Q14'] + dfI1['QI_1'] + dfI3['QI_3'] + dfI7['QI_7']+ dfI9['QI_9'] + dfI10['QI_10']+ dfI12['QI_12']+ dfI13['QI_13']+ dfI14['QI_14']
df_baseLoad
```

### base load profile with industrial load -- Table1

```python
baseLoadProfile = df_baseLoad[['Time', 'Total_P', 'Total_Q']]
baseLoadProfile
```

### Base Load graph

```python
plt.figure(figsize=(10,6))
plt.title('Base Load profile Vs Time')

plt.plot(baseLoadProfile.Time/3600, baseLoadProfile.Total_P/1000,'b', label='BaseLoadProfile')
#plt.plot(newdf.Time/3600, newdf.P1/1000,'b', label='P1')
#plt.plot(newdf.Time/3600, newdf.Q3/1000,'r', label='Q3')
#plt.plot(newdf.Time/3600, newdf.Total_Q/1000, 'r', label='Active_Q')

plt.xlabel('Time (hours)')
plt.ylabel('Load (MW)')

plt.xlim(0,24)
plt.xticks([0,4,8,12,16,20,24])

plt.legend()

plt.savefig('BaseLoadprofile.png', dpi=300)
plt.show()
```

## 1. Only household load -- Table2

```python
df_house = pd.concat([df1, df3, df4, df5, df6, df8, df10, df11, df12, df14], axis=1, join='inner').sort_index()

df_house['P'] = df1['P1'] + df3['P3'] + df4['P4'] + df5['P5'] + df6['P6'] + df8['P8'] + df10['P10'] + df11['P11'] + df12['P12'] + df14['P14']
df_house['Q'] = df1['Q1'] + df3['Q3'] + df4['Q4'] + df5['Q5'] + df6['Q6'] + df8['Q8'] + df10['Q10'] + df11['Q11'] + df12['Q12'] + df14['Q14']
df_house[['Time', 'P', 'Q']]
```

### With 20% EV load to household loads -- Table3

```python
df_house['EV'] = df_house['P'] * 0.2
df_house['Tot_P'] = df_house['P'] + df_house['EV']
df_house[['Time', 'P', 'EV', 'Tot_P']]
```

## 2. Only industrial load -- Table4

```python
df_Industrial = pd.concat([df1, dfI1, dfI3, dfI7, dfI9, dfI10, dfI12, dfI13, dfI14], axis=1, join='inner').sort_index() ## df1 included to get the index

df_Industrial['P_I'] = dfI1['PI_1'] + dfI3['PI_3'] + dfI7['PI_7']+ dfI9['PI_9'] + dfI10['PI_10']+ dfI12['PI_12']+ dfI13['PI_13']+ dfI14['PI_14']
df_Industrial['Q_I'] = dfI1['QI_1'] + dfI3['QI_3'] + dfI7['QI_7']+ dfI9['QI_9'] + dfI10['QI_10']+ dfI12['QI_12']+ dfI13['QI_13']+ dfI14['QI_14']
df_Industrial[['Time', 'P_I', 'Q_I']]
#df_Industrial
```

### Total load profile with 20% EV for Case 2 scenario -- Table5

```python
df_withEVload = pd.concat([df_house, df_Industrial], axis=1, join='inner').sort_index()

df_withEVload['ActiveP']= df_house['Tot_P'] + df_Industrial['P_I']
df_withEVload['ReactiveP'] = df_house['Q'] + df_Industrial['Q_I']

LoadProfileWithEV = df_withEVload[['Time', 'ActiveP', 'ReactiveP']]
LoadProfileWithEV
```

### Graph with load profile with addition of 20% EV load

```python
plt.figure(figsize=(10,6))
plt.title('Load Profile With EV Vs Time')


plt.plot(baseLoadProfile.Time/3600, baseLoadProfile.Total_P/1000,'b', label='BaseLoadProfile')
plt.plot(baseLoadProfile.Time/3600, LoadProfileWithEV.ActiveP/1000,'r', label='LoadProfileWithEV')
#plt.plot(LoadProfileWithEV.Time/3600, LoadProfileWithEV.ActiveP/1000,'b', label ='LoadProfileWithEV')

plt.xlabel('Time (hours)')
plt.ylabel('Load (MW)')

plt.xlim(0,24)
plt.xticks([0,4,8,12,16,20,24])

plt.legend()

plt.savefig('LoadProfileWithEV.png', dpi=300)
plt.show()
```
